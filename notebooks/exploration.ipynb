```python
import requests
import pandas as pd
```

```python
api_secop2 = 'https://www.datos.gov.co/resource/jbjy-vk9h.json'
all_data =[]
offset = 0
limit=1000

while True:
    response = requests.get(api_secop2, params={'$where': 'fecha_de_inicio_del_contrato >= "2024-01-01" AND fecha_de_inicio_del_contrato < "2026-01-01"','$limit': limit, '$offset': offset})
    if response.status_code == 200:
        data = response.json()
        if not data:
            break
        all_data.extend(data)
        offset += limit
    else:
        print(f"Error: {response.status_code}")
        break


df = pd.DataFrame(all_data)
```

#Table dim_date

```python
dim_date = pd.concat([df['fecha_de_inicio_del_contrato'], df['fecha_de_fin_del_contrato'], df['fecha_de_firma'], df['fecha_inicio_liquidacion'], df['fecha_fin_liquidacion']])
```

```python
date = pd.DataFrame(dim_date.drop_duplicates().dropna().reset_index(drop=True), columns=['date'])
```

```python
date['date'] = pd.to_datetime(date['date'], format='ISO8601', errors='coerce')
```

```python
df.loc[df['fecha_de_fin_del_contrato'] >= '2100-01-01T00:00:00.000', ['fecha_de_inicio_del_contrato', 'fecha_de_fin_del_contrato','estado_contrato','nombre_entidad','proveedor_adjudicado','valor_del_contrato']] 
```

```python
date = date.drop_duplicates()
```

```python
date.dropna(inplace=True)
```

```python
date.head()
```

```python
date['year'] = date['date'].dt.year.astype('Int64')
date['month'] = date['date'].dt.month.astype('Int64')
date['day'] = date['date'].dt.day.astype('Int64')
date['quarter'] = date['date'].dt.quarter.astype('Int64')
date['week'] = date['date'].dt.isocalendar().week.astype('Int64')
date['day_of_week'] = date['date'].dt.dayofweek.astype('Int64')
date['day_name'] = date['date'].dt.day_name()
date['month_name'] = date['date'].dt.month_name()
date['is_weekend'] = date['day_of_week'].isin([5, 6])
```

```python
date['id_date'] = date['date'].dt.strftime('%Y%m%d').astype('Int64')
```
