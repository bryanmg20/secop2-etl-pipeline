```python
import requests
import pandas as pd
```

```python
api_secop2 = 'https://www.datos.gov.co/resource/jbjy-vk9h.json'
all_data =[]
offset = 0
limit=50000

while True:
    response = requests.get(api_secop2, params={'$where': 'fecha_de_firma >= "2024-01-01" AND fecha_de_firma < "2025-01-01"','$limit': limit, '$offset': offset})
    if response.status_code == 200:
        data = response.json()
        if not data:
            break
        all_data.extend(data)
        offset += limit
    else:
        print(f"Error: {response.status_code}")
        break


df = pd.DataFrame(all_data)
```

```python
df.to_csv('../data/raw/secop2_data_2024.csv', index=False)
```
